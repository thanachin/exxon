```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(df2_aggregated['OUTBOUND_DATE'], df2_aggregated['NET_QUANTITY_MT'])
plt.xlabel('Outbound Date')
plt.ylabel('Net Quantity (MT)')
plt.title('Time Series of Net Quantity Inbound Over Time')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('Time Series of Net Quantity Outbound Over Time')
plt.show()
```

```python
# Rename the date column in df2_aggregated for merging
df2_aggregated = df2_aggregated.rename(columns={'OUTBOUND_DATE': 'Date'})
df_aggregated = df_aggregated.rename(columns={'INBOUND_DATE': 'Date'})

# Merge the aggregated dataframes on the date column
combined_aggregated_df = pd.merge(df_aggregated, df2_aggregated, on='Date', how='outer', suffixes=('_inbound', '_outbound'))

# Fill NaN values with 0 after merging
combined_aggregated_df = combined_aggregated_df.fillna(0)

# Sort by date
combined_aggregated_df = combined_aggregated_df.sort_values('Date')

display(combined_aggregated_df.head())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(combined_aggregated_df['Date'], combined_aggregated_df['NET_QUANTITY_MT_inbound'], label='Net Quantity Inbound')
plt.plot(combined_aggregated_df['Date'], combined_aggregated_df['NET_QUANTITY_MT_outbound'], label='Net Quantity Outbound')
plt.xlabel('Date')
plt.ylabel('Net Quantity (MT)')
plt.title('Time Series of Net Quantity Inbound and Outbound Over Time')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('Time Series of Net Quantity Inbound and Outbound Over Time')
plt.show()
```
